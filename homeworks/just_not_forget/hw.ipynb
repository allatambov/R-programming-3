## Задание 6

Мы пока не обсуждали детально построение графиков с помощью функции `plot()`, но это задание -- хороший повод познакомиться с базовой графикой в R. 

В векторе `firtree_x` хранятся координаты точек по оси x, а в векторе `firtree_y` хранятся координаты точек по оси y. Можем нанести их на график:

````r
firtree_x <- c(2, 0, 1, 0, 1, 0, 1, 2, 1, 2, 1, 2)
firtree_y <- c(1, 1, 3, 3, 5, 5, 7, 5, 5, 3, 3, 1)

# xlim - границы графика по оси x
# ylim - границы графика по оси y
# col - цвет

plot(firtree_x, firtree_y, xlim = c(0, 5), ylim = c(0, 10), col = "darkgreen")
````

А теперь соединим точки линиями ("o" от *overplotted*, см. *help*) и поменяем маркер для точек (*pch*):

````r
plot(firtree_x, firtree_y, xlim = c(0, 5), ylim = c(0, 10), col = "darkgreen", pch = 19, "o")
````

На построенный график можно наносить точки и линии. Добавим точки в виде снежинок:

````r
plot(firtree_x, firtree_y, xlim = c(0, 5), ylim = c(0, 10), col = "darkgreen", pch = 19, "o")
snow_x <- c(0, 0.5, 1.5, 2.5)
snow_y <- c(9, 7, 9, 7)
points(snow_x, snow_y, col = 'blue', pch = 8)
````
